```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel('/content/Dados_Empresas_Energia.xlsx')

df.head()
```

```python
df.describe()
```

```python
df.info()
```

```python
# RESETAR INDEX:

df.set_index('Data', inplace=True)
```

```python
# GRÁFICO:

plt.style.use('seaborn-darkgrid')
plt.figure(figsize=(16,7))
plt.title('Análise ações de empresas de Energia', loc='left', fontsize=18, fontweight=0)

# plot petrobras:
plt.plot(df.index, df['Petrobras'], color='#008c4a', linewidth=4, alpha=0.7)
plt.text(df.index[-1], df['Petrobras'].tail(1), 'Petrobrás', color='#008c4a', size='large', horizontalalignment='left')

# plot de todas as colunas:
for Coluna in df.columns[1:]:
  plt.plot(df.index, df[Coluna], color='gray', linewidth=2, alpha=0.7)
  plt.text(df.index[-1], df[Coluna].tail(1), Coluna, color='gray', size='large', horizontalalignment='left')

# labes:
plt.xlabel('Periodo')
plt.ylabel('Preço Fechamento (R$)')
```
